6) What is the definition of a 180-day moving average?


* Average of last 180 from the present day is called 180 day moving average
* To get the exact curve moving average should be small

```python
hdfc_last_year['Moving_average_180']=hdfc_last_year['Close'].rolling(window=180).mean()
```

```python
hdfc_last_year['Moving_average_9']=hdfc_last_year['Close'].rolling(window=9).mean()
```

```python
hdfc_last_year
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_style("darkgrid");
plt.figure(figsize=(12,8));
plt.rcParams['font.size'] = '16'
plt.plot(hdfc_last_year.Date,hdfc_last_year.Close)
plt.plot(hdfc_last_year.Date,hdfc_last_year.Moving_average_180)
plt.plot(hdfc_last_year.Date,hdfc_last_year.Moving_average_9)
plt.xlabel("Date")
plt.ylabel("Price")
plt.legend(['Close','Moving_average_180','Moving_average_9'])
plt.title("HDFC BANK CLOSING PRICE Vs MOVING AVERAGES");
```

```python
import plotly.express as px
```

```python
fig=px.line(hdfc_last_year,x="Date",y="Close",title="Closing price vs Moving Average")
fig.add_scatter(x=hdfc_last_year['Date'],y=hdfc_last_year['Moving_average_180'])
fig.add_scatter(x=hdfc_last_year['Date'],y=hdfc_last_year['Moving_average_9'])
fig.show()
```
